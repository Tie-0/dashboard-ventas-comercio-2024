```python
import pandas as pd
import numpy as np

np.random.seed(42)

productos = ['Remera', 'Pantalón', 'Zapatillas', 'Campera', 'Gorra', 'Medias', 'Buzo']
categorias = {'Remera': 'Tops', 'Pantalón': 'Bottoms', 'Zapatillas': 'Calzado',
              'Campera': 'Abrigos', 'Gorra': 'Accesorios', 'Medias': 'Accesorios', 'Buzo': 'Tops'}
precios = {'Remera': 5000, 'Pantalón': 12000, 'Zapatillas': 35000,
           'Campera': 25000, 'Gorra': 3500, 'Medias': 1500, 'Buzo': 15000}

n = 500
producto = np.random.choice(productos, n)

df = pd.DataFrame({
    'fecha': pd.date_range(start='2024-01-01', periods=n, freq='D').strftime('%Y-%m-%d')[:n],
    'producto': producto,
    'categoria': [categorias[p] for p in producto],
    'cantidad': np.random.randint(1, 6, n),
    'precio_unitario': [precios[p] for p in producto],
    'ciudad': np.random.choice(['Buenos Aires', 'Córdoba', 'Rosario', 'Mendoza'], n)
})

df['fecha'] = pd.to_datetime(df['fecha'])
df['fecha'] = df['fecha'].dt.normalize() + pd.to_timedelta(np.random.randint(0, 365, n), unit='D')
df['fecha'] = df['fecha'].dt.strftime('%Y-%m-%d')
df['ingreso_total'] = df['cantidad'] * df['precio_unitario']

print(df.head(10))
print(f"\nDataset: {len(df)} ventas generadas")
```
